### 3-1. 서울시 규칙적인 운동 실천 비율 통계 2017

```python
exercise_raw = pd.read_csv("서울시 규칙적인 운동 실천 비율 통계 2017.txt", header=0,sep='\t',encoding='utf-8', names = ['기간', '대분류','분류', '주 3회 이상 규칙적 운동을 한다','주 1~2회 정도 규칙적 운동을 한다','규칙적은 아니지만 운동을 한다','운동을 거의 하지 않는다'])
exercise = exercise_raw[28:]
exercise = exercise.reset_index(drop=True).drop('대분류',axis=1).rename(columns= {'분류':'지역'})
exercise = exercise.drop('기간', axis = 1)
exercise
```

### 3-2. 서울시 삶의 만족 통계 2017

```python
life_raw = pd.read_csv("서울시 삶의 만족 통계 2017.txt", header=0,sep='\t',encoding='utf-8', names = ['기간', '대분류','분류', '종합','생활수준','성취한 것','개인적 인간관계', '체감 안전지역','지역사회 일원','미래의 안정성','좋아하는 것을 하는데 활용하는 시간','지역의 생활환경'] )
life = life_raw[28:]
life = life.reset_index(drop=True).drop('대분류',axis=1).rename(columns= {'분류':'지역'})
life = life.drop('기간', axis = 1)
life
```

### 3-3. 서울시 운동을 하지 않는 이유 통계 2017

```python
hate_raw = pd.read_csv("서울시 운동을 하지 않는 이유 통계 2017.txt", header=0,sep='\t',encoding='utf-8', names = ['기간', '대분류','분류', '운동을 할 충분한 시간이 없어서','함께 운동을 할 사람이 없어서','운동을 할 만한 장소가 없어서','운동을 싫어해서', '기타'] )
hate = hate_raw[28:] 
hate = hate.reset_index(drop=True).drop('대분류',axis=1).rename(columns= {'분류':'지역'})
hate = hate.drop(['운동을 할 충분한 시간이 없어서','함께 운동을 할 사람이 없어서','운동을 싫어해서','기타'], axis=1)
hate = hate.drop('기간', axis = 1)
hate
```

### 3-4. 서울시 주로 이용하는 운동 장소 통계 2017

```python
eplace_raw = pd.read_csv("서울시 주로 이용하는 운동 장소 통계 2017.txt", header=0,sep='\t',encoding='utf-8', names = ['기간', '대분류','분류', '공원 및 산 등의 산책로','민간상업 체육시설','학교체육시설','집에서 간단한 운동', '지자체 및 공공기관운영 시설','직장내 체육시설','기타'] )
eplace = eplace_raw[28:]
eplace = eplace.reset_index(drop=True).drop('대분류',axis=1).rename(columns= {'분류':'지역'})
eplace['기타'] = eplace['기타'].apply(lambda x: x.replace('-','0'))
eplace['민간상업 체육시설'] = eplace['민간상업 체육시설'].astype('float')
eplace['학교체육시설'] = eplace['학교체육시설'].astype('float')
eplace['집에서 간단한 운동'] = eplace['집에서 간단한 운동'].astype('float')
eplace['지자체 및 공공기관운영 시설'] = eplace['지자체 및 공공기관운영 시설'].astype('float')
eplace['직장내 체육시설'] = eplace['직장내 체육시설'].astype('float')
eplace['기타'] = eplace['기타'].astype('float')
eplace['기타장소'] = eplace['민간상업 체육시설'] + eplace['학교체육시설'] + eplace['집에서 간단한 운동'] + eplace['지자체 및 공공기관운영 시설'] + eplace['직장내 체육시설'] + eplace['기타']
eplace = eplace.drop(['민간상업 체육시설', '학교체육시설','집에서 간단한 운동','지자체 및 공공기관운영 시설','직장내 체육시설','기타'], axis = 1)
eplace = eplace.drop('기간', axis = 1)
eplace
```

### 3-5. 서울시 지역 녹지 만족도 2018

```python
green_raw = pd.read_csv("서울시 지역 녹지 만족도 2018.txt", header=0,sep='\t',encoding='utf-8', names = ['기간', '대분류','분류', '계','매우 불만족','약간 불만족','보통', '약간 만족','매우 만족'] )
green = green_raw[28:]
green = green.reset_index(drop=True).drop('대분류',axis=1).rename(columns= {'분류':'지역'})
green['만족응답'] = green['약간 만족'] + green['매우 만족']
green = green.drop(['기간','계','매우 불만족','약간 불만족','보통','약간 만족','매우 만족'], axis = 1)
green
```

### 3-6. 서울시 자전거 이용률 통계 2018

```python
ride_raw = pd.read_csv("서울시 자전거 이용률 통계 2018.txt", header=0,sep='\t',encoding='utf-8', names = ['기간', '대분류','분류', '합계','이동 수단으로 이용','운동 수단으로 이용','이용하지 않음'] )
ride = ride_raw[28:]
ride['이용비율'] = ride['이동 수단으로 이용'] + ride['운동 수단으로 이용']
ride = ride.drop(['기간','합계','이동 수단으로 이용','운동 수단으로 이용','이용하지 않음'], axis = 1)
ride = ride.reset_index(drop=True).drop('대분류',axis=1).rename(columns= {'분류':'지역'})
ride
```

```python
X = X.set_index(['지역'])
```

```python
X = X.dropna()
```

### feature vector 표준화

```python
from sklearn.preprocessing import StandardScaler

x_std = StandardScaler().fit_transform(X)

x_std
```

```python
# Y is labels

Y = test[['지역']]
Y.head()
```

### scikit-learn 활용 PCA

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=3)  # 주성분 분석의 축 갯수를 의미. 1이면 1개의 차원 1개축으로, 2면 2개차원 2개축으로
sklearn_pca_x = pca.fit_transform(x_std)
sklearn_pca_x
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sklearn_result = pd.DataFrame(sklearn_pca_x, columns=['PC1','PC2','PC3'])
sklearn_result['y-axis'] = 0.0
sklearn_result['label'] = Y
sns.lmplot('PC2', 'PC3', data=sklearn_result, fit_reg=False, # x-axis, y-axis, data, no line
    scatter_kws={"s": 50}, # marker size
    hue="label") # color

plt.xlabel('주성분1')
plt.ylabel('주성분2')
```
